```python
import geopandas as gpd
from geopandas import GeoSeries, GeoDataFrame
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
education_1 = gpd.read_file('amenities_shapefile/education/education_1.shp')
education_2 = gpd.read_file('amenities_shapefile/education/education_2.shp')
health_1 = gpd.read_file('amenities_shapefile/health/health_1.shp')
health_2 = gpd.read_file('amenities_shapefile/health/health_2.shp')
finance_1 = gpd.read_file('amenities_shapefile/finance/finance_1.shp')
finance_2 = gpd.read_file('amenities_shapefile/finance/finance_2.shp')
government_1 = gpd.read_file('amenities_shapefile/government/government_1.shp')
government_2 = gpd.read_file('amenities_shapefile/government/government_2.shp')
groceries_1 = gpd.read_file('amenities_shapefile/groceries/groceries_1.shp')
groceries_2 = gpd.read_file('amenities_shapefile/groceries/groceries_2.shp')
mall_1 = gpd.read_file('amenities_shapefile/malls/mall_1.shp')
mall_2 = gpd.read_file('amenities_shapefile/malls/mall_2.shp')
residence_1 = gpd.read_file('amenities_shapefile/residence/residence_1.shp')
residence_2 = gpd.read_file('amenities_shapefile/residence/residence_2.shp')
security_1 = gpd.read_file('amenities_shapefile/security/security_1.shp')
security_2 = gpd.read_file('amenities_shapefile/security/security_2.shp')
transportation = gpd.read_file('amenities_shapefile/transportation/transportation.shp')
```

```python
education_merge = education_1.merge(education_2, how='outer')
health_merge = health_1.merge(health_2, how='outer')
finance_merge = finance_1.merge(finance_2, how='outer')
government_merge = government_1.merge(government_2, how='outer')
groceries_merge = groceries_1.merge(groceries_2, how='outer')
mall_merge = mall_1.merge(mall_2, how='outer')
residence_merge = residence_1.merge(residence_2, how='outer')
security_merge = security_1.merge(security_2, how='outer')

```

```python
amenity_merge = education_merge.merge(health_merge, how='outer')
amenity_merge = amenity_merge.merge(finance_merge, how='outer')
amenity_merge = amenity_merge.merge(government_merge, how='outer')
amenity_merge = amenity_merge.merge(groceries_merge, how='outer')
amenity_merge = amenity_merge.merge(mall_merge, how='outer')
amenity_merge = amenity_merge.merge(residence_merge, how='outer')
amenity_merge = amenity_merge.merge(security_merge, how='outer')
amenity_merge = amenity_merge.merge(transportation, how='outer')
```

```python
amenity_colors = {'education': 'red', 'health': 'blue', 'finance': 'green', 'government offices': 'orange', 'grocery': 'purple', 'malls': 'yellow', 'residential areas': 'pink', 'security':'black', 'transportation':'white'}
cmap = ListedColormap([amenity_colors[amenity] for amenity in amenity_merge['amenity'].unique()])
```

```python
amenity_merge.plot(column='amenity', cmap=cmap, legend=True, figsize=(40, 40))
```
